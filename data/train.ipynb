```python


from ultralytics import YOLO
import os

# === PATH CONFIG (LOCAL WINDOWS) ===
DATASET_ROOT = r"D:\digital-image-processing\StoreItemDetection\data\yolo_dataset"

print("Dataset root:", DATASET_ROOT)
print("Train images:", len(os.listdir(os.path.join(DATASET_ROOT, "images", "train"))))
print("Val images:", len(os.listdir(os.path.join(DATASET_ROOT, "images", "val"))))

# Load YOLO model
model = YOLO("yolov8n.pt")   # small, fast (optional: yolov8s.pt, yolov8m.pt ...)

# === TRAIN YOLO ===
model.train(
    data=os.path.join(DATASET_ROOT, "dataset.yaml"),
    epochs=100,
    imgsz=640,
    batch=16,
    name="train_yolo_local",
    project=r"D:\digital-image-processing\StoreItemDetection\runs"
)

# === BEST MODEL PATH ===
best_weight = r"D:\digital-image-processing\StoreItemDetection\runs\detect\train_yolo_local\weights\best.pt"
print("\nBEST MODEL SAVED AT:", best_weight)
```
